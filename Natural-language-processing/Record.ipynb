```python
import numpy as np
import pyaudio
import matplotlib.pyplot as plt
import warnings
```

```python
sample = pyaudio.PyAudio()
stream = sample.open(format=pyaudio.paInt16,channels=1,rate=44100,input=True,frames_per_buffer=1024)
```

```python
def record():
   # warnings.simplefilter("ignore")
    data = stream.read(1024*220)
    decoded = np.fromstring(data, dtype='int32')
    trigger = np.argmax(np.abs(decoded) > 0.*10**9)
    return decoded[trigger:trigger+24100]
```

```python
#Up
for i in range(5):
    print('please say "Up"')
    np.save(f"recordings/Up{i}",record())
```

```python
#Down
for i in range(5):
    print('please say "Down"')
    np.save(f"recordings/Down{i}",record())
```
